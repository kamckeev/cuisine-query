```python
import pandas as pd
```

```python
file="data/train.json"
df = pd.read_json(file, orient='columns')
df.head()
```

```python
df['ing_string']=df['ingredients'].astype(str)
df['ing_string']=df['ing_string'].replace('[', "'", regex=True)
df.head()
```

```python
ingredients = df['ingredients']
all_ingredients = []
for item in ingredients:
    all_ingredients.extend(item)
    
#This list will need some cleanup. whole lotta cleanup for data visualization
unique_ingredients = list(set(all_ingredients))
# unique_ingredients
```

```python
print("all_ingredients: " + str(len(all_ingredients)))
print("unique_ingredients: " + str(len(unique_ingredients)))
```

```python
cuisines = df.groupby('cuisine').count()

cuisine_list = list(set(df['cuisine']))
cuisine_list
# cuisines
```

```python
cuisine_list
big_dict={}
for item in cuisine_list:
    big_dict[item]=[]

big_dict
```

```python
for index, row in df.iterrows():
    big_dict[row['cuisine']].append(row['ingredients'][0])
big_dict
```

```python
num_dict={}
for item in cuisine_list:
    num_dict[item]={}

num_dict
```

```python
for item in cuisine_list:
    list=big_dict[item]
    for ingred in list:
        if ingred in num_dict[item]:
            num_dict[item][ingred]+=1
        else:
            num_dict[item][ingred]=1
print (num_dict)
```
